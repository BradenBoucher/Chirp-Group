```python
print('First Script. Yay')

import gdsfactory as gf 
import random

angle = random.randint(0, 360)  # Generates an integer between 0 and 90 (inclusive)
#print(angle)

gf.clear_cache()


c = gf.Component("text_with_rotation")


Chirp = gf.components.text(
    text="Chirp",
    size=10,
    layer=(1, 0)
)

text = c.add_ref(Chirp)
text.rotate(angle)

c.plot()


```
